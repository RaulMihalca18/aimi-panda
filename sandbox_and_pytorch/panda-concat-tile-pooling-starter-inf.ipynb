# Model

```python
class AdaptiveConcatPool2d(nn.Module):
    def __init__(self, size=None):
        super(AdaptiveConcatPool2d, self).__init__()
        self.avg_pool = nn.AdaptiveAvgPool2d(size)
        self.max_pool = nn.AdaptiveMaxPool2d(size)
        
    def forward(self, x):
        x_avg = self.avg_pool(x)
        x_max = self.max_pool(x)
        return torch.cat([x_avg, x_max], dim=1)

class Flatten(nn.Module):
    def forward(self, x):
        return x.view(x.size(0), -1)
    
def _resnext(url, block, layers, pretrained, progress, **kwargs):
    model = ResNet(block, layers, **kwargs)
    #state_dict = load_state_dict_from_url(url, progress=progress)
    #model.load_state_dict(state_dict)
    return model

class Model(nn.Module):
    def __init__(self, arch='resnext50_32x4d', n=6, pre=True):
        super().__init__()
        #m = torch.hub.load('facebookresearch/semi-supervised-ImageNet1K-models', arch)
        m = _resnext(semi_supervised_model_urls[arch], Bottleneck, [3, 4, 6, 3], False, 
                progress=False,groups=32,width_per_group=4)
        self.enc = nn.Sequential(*list(m.children())[:-2])       
        nc = list(m.children())[-1].in_features
        self.head = nn.Sequential(AdaptiveConcatPool2d(),Flatten(),nn.Linear(2*nc,512),
                Mish(),nn.BatchNorm1d(512),nn.Dropout(0.5),nn.Linear(512,n))
        
    def forward(self, x):
        shape = x.shape
        n = shape[1]
        x = x.view(-1,shape[2],shape[3],shape[4])
        x = self.enc(x)
        shape = x.shape
        x = x.view(-1,n,shape[1],shape[2],shape[3]).permute(0,2,1,3,4).contiguous()\
          .view(-1,shape[1],shape[2]*n,shape[3])
        x = self.head(x)
        return x
```

```python
models = []
for path in MODELS:
    state_dict = torch.load(path,map_location=torch.device('cpu'))
    model = Model()
    model.load_state_dict(state_dict)
    model.float()
    model.eval()
    model.cuda()
    models.append(model)

del state_dict
```

# Data

```python
from torch.utils.data import Dataset
from torch.utils.data import DataLoader

def tile(img):
    shape = img.shape
    pad0, pad1 = (sz - shape[0] % sz) % sz, (sz - shape[1] % sz) % sz
    img = np.pad(img, [[pad0 // 2, pad0 - pad0 // 2], [pad1 // 2, pad1 - pad1 // 2], [0, 0]], constant_values=255)
    img = img.reshape(img.shape[0] // sz, sz, img.shape[1] // sz, sz, 3)
    img = img.transpose(0, 2, 1, 3, 4).reshape(-1, sz, sz, 3)
    if len(img) < N:
        img = np.pad(img, [[0, N - len(img)], [0, 0], [0, 0], [0, 0]], constant_values=255)
    idxs = np.argsort(img.reshape(img.shape[0], -1).sum(-1))[:N]
    img = img[idxs]
    return img

mean = torch.tensor([1.0 - 0.90949707, 1.0 - 0.8188697, 1.0 - 0.87795304])
std = torch.tensor([0.36357649, 0.49984502, 0.40477625])

class PandaDataset(Dataset):
    def __init__(self, path, test):
        self.path = path
        self.names = list(pd.read_csv(test).image_id)

    def __len__(self):
        return len(self.names)

    def __getitem__(self, idx):
        name = self.names[idx]
        img = skimage.io.MultiImage(os.path.join(DATA, name + '.tiff'))[-1]
        tiles = torch.Tensor(1.0 - tile(img) / 255.0)
        tiles = (tiles - mean) / std
        return tiles.permute(0, 3, 1, 2), name

    def get_dataloader(self, batch_size=32, shuffle=True, num_workers=0):
        return DataLoader(self, batch_size=batch_size, shuffle=shuffle, num_workers=num_workers)
```

# Prediction

```python
# Assuming you have defined the PandaDataset class as mentioned in the previous response

DATA = '../input/prostate-cancer-grade-assessment/train_images'
TEST = '../input/prostate-cancer-grade-assessment/train.csv'
SAMPLE = '../input/prostate-cancer-grade-assessment/sample_submission.csv'

# Load sample DataFrame
sub_df = pd.read_csv(SAMPLE)

# Check if data directory exists
if os.path.exists(DATA):
    # Create PandaDataset instance
    ds = PandaDataset(DATA, TEST)
    # Create DataLoader
    dl = ds.get_dataloader(batch_size=bs, num_workers=nworkers, shuffle=False)
    names, preds = [], []

    # Iterate over DataLoader
    with torch.no_grad():
        for x, y in tqdm(dl):
            # Move input to GPU if available
            x = x.cuda()
            # Apply dihedral TTA
            x = torch.stack([
                x, x.flip(-1), x.flip(-2), x.flip(-1, -2),
                x.transpose(-1, -2), x.transpose(-1, -2).flip(-1),
                x.transpose(-1, -2).flip(-2), x.transpose(-1, -2).flip(-1, -2)
            ], 1)
            x = x.view(-1, N, 3, sz, sz)
            # Get predictions from model
            p = [model(x) for model in models]
            p = torch.stack(p, 1)
            p = p.view(bs, 8 * len(models), -1).mean(1).argmax(-1).cpu()
            # Append names and predictions
            names.append(y)
            preds.append(p)

    # Concatenate names and predictions
    names = np.concatenate(names)
    preds = torch.cat(preds).numpy()
    # Create DataFrame from predictions
    sub_df = pd.DataFrame({'image_id': names, 'isup_grade': preds})
    # Save DataFrame to CSV
    sub_df.to_csv('submission.csv', index=False)
    # Display DataFrame head
    print(sub_df.head())
```
